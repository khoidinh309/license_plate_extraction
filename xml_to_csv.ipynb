```python
import pandas as pd
import xml.etree.ElementTree as xet
```

```python
from glob import glob
```

```python
path = glob('./images/*.xml')
path
```

```python
labels_dicts = dict(filepath=[], xmin=[], xmax=[], ymin=[], ymax=[])
for filename in path:
    labels_dicts['filepath'].append(filename)
    info = xet.parse(filename)
    root = info.getroot()
    member_object = root.find('object')
    labels_info = member_object.find('bndbox')
    xmin = int(labels_info.find('xmin').text)
    xmax = int(labels_info.find('xmax').text)
    ymin = int(labels_info.find('ymin').text)
    ymax = int(labels_info.find('ymax').text)
    labels_dicts['xmin'].append(xmin)
    labels_dicts['xmax'].append(xmax)
    labels_dicts['ymin'].append(ymin)
    labels_dicts['ymax'].append(ymax)
    
labels_dicts
```

```python
df = pd.DataFrame(labels_dicts)
df
```

```python
df.to_csv('labels.csv', index=False)
```
